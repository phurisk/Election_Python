##1. How many movies have an average rating over 4.5 overall?

```python
average_ratings = merged_df.groupby('Title')['Rating'].mean()
movies_over_45 = average_ratings[average_ratings > 4.5]
num_movies_over_45 = len(movies_over_45)
print("Number of movies with an average rating over 4.5 overall:", num_movies_over_45)
```

#2. How many movies have an average rating over 4.5 among men? How about women?

```python

# กรอง DataFrame เพื่อรวมเฉพาะการให้คะแนนโดยผู้ชาย
average_ratings_men = merged_df[merged_df['Gender'] == 'M'].groupby('Title')['Rating'].mean()

# กรอง DataFrame เพื่อรวมเฉพาะการให้คะแนนโดยผู้หญิง
average_ratings_women = merged_df[merged_df['Gender'] == 'F'].groupby('Title')['Rating'].mean()

# นับจำนวนหนังที่มีเรตติ้งเฉลี่ยมากกว่า 4.5 ทั้งชายและหญิง
num_movies_over_45_men = len(average_ratings_men[average_ratings_men > 4.5])
num_movies_over_45_women = len(average_ratings_women[average_ratings_women > 4.5])

print("Number of movies with an average rating over 4.5 among men:", num_movies_over_45_men)
print("Number of movies with an average rating over 4.5 among women:", num_movies_over_45_women)
```

#3. How many movies have an median rating over 4.5 among men over age 30? How about women over age 30?

```python


# กรอง DataFrame เพื่อรวมเฉพาะการให้คะแนนโดยผู้ชายที่มีอายุเกิน 30 ปี
median_ratings_men_over_30 = merged_df[(merged_df['Gender'] == 'M') & (merged_df['Age'] > 30)].groupby('Title')['Rating'].median()

# กรอง DataFrame เพื่อรวมเฉพาะการให้คะแนนโดยผู้หญิงที่มีอายุเกิน 30 ปี
median_ratings_women_over_30 = merged_df[(merged_df['Gender'] == 'F') & (merged_df['Age'] > 30)].groupby('Title')['Rating'].median()

# นับจำนวนภาพยนตร์ที่มีเรตติ้งเฉลี่ยมากกว่า 4.5 สำหรับชายและหญิงที่มีอายุมากกว่า 30 ปี
num_movies_over_45_median_men = len(median_ratings_men_over_30[median_ratings_men_over_30 > 4.5])
num_movies_over_45_median_women = len(median_ratings_women_over_30[median_ratings_women_over_30 > 4.5])

print("Number of movies with a median rating over 4.5 among men over age 30:", num_movies_over_45_median_men)
print("Number of movies with a median rating over 4.5 among women over age 30:", num_movies_over_45_median_women)
```

```python
import pandas as pd

# คำนวณความถี่ของชื่อหนังแต่ละเรื่อง
movie_frequency = merged_df['Title'].value_counts().reset_index()
movie_frequency.columns = ['Title', 'Frequency']

# Calculate the mean rating for each movie title
movie_ratings = merged_df.groupby('Title')['Rating'].mean().reset_index()
movie_ratings.columns = ['Title', 'Mean Rating']

# Merge the two DataFrames on the 'Title' column
merged_data = pd.merge(movie_frequency, movie_ratings, on='Title')

# Print the top 10 movies by frequency and their mean ratings
top_10_movies = merged_data.sort_values(by='Frequency', ascending=False).head(10)
print("Top 10 movies by frequency and their mean ratings:")
print(top_10_movies)
```

#5. Make some conjectures about how easy various groups are to please? Support your answers with data!

```python
age_mapping = {
    1: 'Under 18',
    18: '18-24',
    25: '25-34',
    35: '35-44',
    45: '45-49',
    50: '50-55',
    56: '56+'
}

occupation_mapping = {
    0: 'other',
    1: 'academic/educator',
    2: 'artist',
    3: 'clerical/admin',
    4: 'college/grad student',
    5: 'customer service',
    6: 'doctor/health care',
    7: 'executive/managerial',
    8: 'farmer',
    9: 'homemaker',
    10: 'K-12 student',
    11: 'lawyer',
    12: 'programmer',
    13: 'retired',
    14: 'sales/marketing',
    15: 'scientist',
    16: 'self-employed',
    17: 'technician/engineer',
    18: 'tradesman/craftsman',
    19: 'unemployed',
    20: 'writer'
}

# Update the 'Age' and 'Occupation' columns using the mappings
merged_df['Age'] = merged_df['Age'].map(age_mapping)
merged_df['Occupation'] = merged_df['Occupation'].map(occupation_mapping)
```

อาชีพนักวิทยาศาสตร์ จะมี mean rating กับหนังประเภท Sci-Fi มากกว่าอาชีพอื่นๆ

```python
# Filter the DataFrame for individuals with the occupation 'Scientist'
scientist_ratings = merged_df[merged_df['Occupation'] == 'scientist']

# Filter the DataFrame for Sci-Fi movies
scientist_scifi_ratings = scientist_ratings[scientist_ratings['Genre'].str.contains('Sci-Fi')]

# Calculate the average rating for Sci-Fi movies by Scientists
avg_rating_scientist_scifi = scientist_scifi_ratings['Rating'].mean()

# Calculate the average rating for Sci-Fi movies for all other occupations
other_occupations_data = merged_df[merged_df['Occupation'] != 'Scientist']
sci_fi_movies_other_occupations = other_occupations_data[other_occupations_data['Genre'].str.contains('Sci-Fi')]
avg_rating_sci_fi_other_occupations = sci_fi_movies_other_occupations['Rating'].mean()

# Print the results
print("Average rating for Sci-Fi movies by Scientists:", avg_rating_scientist_scifi)
print("Average rating for Sci-Fi movies by all other occupations:", avg_rating_sci_fi_other_occupations)
```

```python
# Calculate the average rating for Sci-Fi movies by each occupation
occupation_ratings = {}
for occupation in merged_df['Occupation'].unique():
    occupation_data = merged_df[merged_df['Occupation'] == occupation]
    sci_fi_movies = occupation_data[occupation_data['Genre'].str.contains('Sci-Fi')]
    avg_rating = sci_fi_movies['Rating'].mean()
    occupation_ratings[occupation] = avg_rating

# Convert the occupation_ratings dictionary to a DataFrame for easier display
occupation_ratings_df = pd.DataFrame(list(occupation_ratings.items()), columns=['Occupation', 'Average Rating'])

# Sort the DataFrame by average rating in descending order
occupation_ratings_df = occupation_ratings_df.sort_values(by='Average Rating', ascending=False)

# Print the results
print("Average rating for Sci-Fi movies by each occupation:")
print(occupation_ratings_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the occupation_ratings data in a DataFrame
# Create a DataFrame from the occupation_ratings dictionary
occupation_ratings_df = pd.DataFrame(list(occupation_ratings.items()), columns=['Occupation', 'Average Rating'])

# Sort the DataFrame by 'Average Rating' in descending order
top_5_occupations = occupation_ratings_df.sort_values(by='Average Rating', ascending=False).head(5)

# Define custom colors
custom_colors = ['#2E86C1', '#17A589', '#E67E22', '#95A5A6', '#884EA0']

# Create a bar plot for the top 5 occupations with detailed scale and rating values
plt.figure(figsize=(10, 6))
bars = plt.bar(top_5_occupations['Occupation'], top_5_occupations['Average Rating'], color=custom_colors)
plt.xlabel('Occupation')
plt.ylabel('Average Rating')
plt.title('Top 5 Occupations with Highest Average Rating for Sci-Fi Movies')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Annotate the rating values on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 2), textcoords='offset points',
                 ha='center', va='bottom')

plt.tight_layout()

# Show the plot
plt.show()
```

### A scatter plot of men versus women and their mean rating for every movie

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have the data in a DataFrame called 'merged_df'

# Filter the data to include only rows where 'Gender' is 'M' (for men)
men_df = merged_df[merged_df['Gender'] == 'M']

# Filter the data to include only rows where 'Gender' is 'F' (for women)
women_df = merged_df[merged_df['Gender'] == 'F']

# Group by 'MovieID' and calculate the mean rating for men and women separately
men_ratings = men_df.groupby('MovieID')['Rating'].mean()
women_ratings = women_df.groupby('MovieID')['Rating'].mean()

# Merge the men_ratings and women_ratings DataFrames based on 'MovieID'
ratings_comparison = pd.merge(men_ratings, women_ratings, on='MovieID', suffixes=('_Men', '_Women'))

# Create a scatter plot of mean ratings for men versus women for all movies
plt.figure(figsize=(10, 6))
cmap = plt.get_cmap('coolwarm', len(ratings_comparison))  # Choose a colormap with as many colors as data points
colors = [cmap(i) for i in range(len(ratings_comparison))]
plt.scatter(ratings_comparison['Rating_Men'], ratings_comparison['Rating_Women'], alpha=0.5, c=colors)
plt.xlabel("Mean Rating for Men")
plt.ylabel("Mean Rating for Women")
plt.title("Mean Ratings for Men vs. Women (All Movies)")
plt.grid()

# Create a color legend based on gender
handles = [plt.Line2D([0], [0], marker='o', color='w', label='Men', markerfacecolor='blue', markersize=10),
           plt.Line2D([0], [0], marker='o', color='w', label='Women', markerfacecolor='red', markersize=10)]
plt.legend(handles=handles)

plt.show()
print("Number of rows for men:", len(men_ratings))
print("Number of rows for women:", len(women_ratings))
```

### A scatter plot of men versus women and their mean rating for movies rated more than 200 times

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the data in a DataFrame called 'merged_df'

# Filter movies with more than 200 ratings
popular_movies = movie_ratings_count[movie_ratings_count > 200].index
popular_movie_avg_ratings = movie_avg_ratings[popular_movies]

# Filter the data to include only rows where 'Gender' is 'M' (for men) for popular movies
men200_df = merged_df[(merged_df['Gender'] == 'M') & (merged_df['MovieID'].isin(popular_movies))]

# Filter the data to include only rows where 'Gender' is 'F' (for women) for popular movies
women200_df = merged_df[(merged_df['Gender'] == 'F') & (merged_df['MovieID'].isin(popular_movies))]

# Group by 'MovieID' and calculate the mean rating for men and women separately for popular movies
men_ratings = men200_df.groupby('MovieID')['Rating'].mean()
women_ratings = women200_df.groupby('MovieID')['Rating'].mean()

# Create a scatter plot of mean ratings for men versus women for popular movies
plt.figure(figsize=(10, 6))
plt.scatter(men_ratings, women_ratings, alpha=0.5, c='blue', label='Men vs. Women', edgecolor='purple')
plt.xlabel("Mean Rating for Men (Popular Movies)")
plt.ylabel("Mean Rating for Women (Popular Movies)")
plt.title("Mean Ratings for Men vs. Women (Movies Rated > 200 times)")
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats  # Import the stats module from SciPy

# Define the specific occupations
specific_occupations = ['scientist', 'technician/engineer', 'programmer', 'doctor/health care']

# Filter the data for Sci-Fi movies and specific occupations
filtered_df = merged_df[(merged_df['Genre'] == 'Sci-Fi') & (merged_df['Occupation'].isin(specific_occupations))]

# Calculate mean ratings for men and women
men_ratings = filtered_df[filtered_df['Gender'] == 'M']['Rating']
women_ratings = filtered_df[filtered_df['Gender'] == 'F']['Rating']

# Perform a statistical test to compare the ratings between men and women
t_stat, p_value = stats.ttest_ind(men_ratings, women_ratings)

# Define a significance level (e.g., 0.05)
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    significance = "significant"
else:
    significance = "not significant"

# Create a bar chart to show the mean ratings for men and women
plt.figure(figsize=(8, 6))
labels = ['μ_Men', 'μ_Women']
results = [men_ratings.mean(), women_ratings.mean()]
plt.bar(labels, results, color=['blue', '#DE3163'])
plt.xlabel('Gender')
plt.ylabel('Mean Rating')
plt.title(f'Mean Ratings for Sci-Fi Movies by Gender (Specific Occupations: {", ".join(specific_occupations)})')
plt.ylim(0, 5)  # Set the y-axis range from 0 to 5
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the p-value and significance as text on the chart
plt.text(0.5, 4.8, f'p-value: {p_value:.4f}', fontsize=12, ha='center')
plt.text(0.5, 4.6, f'Significance: {significance}', fontsize=12, ha='center')

# Save or display the chart
plt.tight_layout()
plt.savefig('rating_gender_occupations.png')  # Save the chart as an image
plt.show()
```
