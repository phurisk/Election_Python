#คะแนนรวมปาตี้ลิส (1)

- คะแนนดิบ

```python
import pandas as pd

# Merge result_constituencies_Candidate with info_party_overview on 'party_id' and 'id'
merged_df = result_constituencies_Candidate[['party_id', 'mp_app_rank']].merge(
    info_party_overview[['id', 'name']], left_on='party_id', right_on='id', how='left'
)

# Select the desired columns
final_merged_df = merged_df[['party_id', 'mp_app_rank', 'name']]

# Filter rows where 'mp_app_rank' is equal to 1
rank_counts = final_merged_df[final_merged_df['mp_app_rank'] == 1]

# Group by 'name' and count the occurrences of '1'
rank_counts = rank_counts.groupby('name')['mp_app_rank'].count().reset_index()

# Rename the count column
rank_counts = rank_counts.rename(columns={'mp_app_rank': 'raw_score'})

# Print the final DataFrame with counts of '1'

rank_counts
```

- คะแนนปารตี้ลิสยังไม่รวม(คะแนนแค่ละพรรคที่ได้)

```python
import pandas as pd

# Merge result_constituencies_PartyList with info_party_overview on 'party_id'
merged_df = result_constituencies_PartyList[['party_id', 'party_list_vote']].merge(
    info_party_overview[['id', 'name']], left_on='party_id', right_on='id', how='inner'
)

# Select only the desired columns
final_merged_df = merged_df[['party_id', 'party_list_vote', 'name']]

# Group by 'name' and calculate the sum of 'party_list_vote' for each group
party_vote_totals = final_merged_df.groupby('name')['party_list_vote'].sum().reset_index()

# Rename the 'party_list_vote' column to 'total_party_list_vote' if needed
party_vote_totals.rename(columns={'party_list_vote': 'total_party_list_vote'}, inplace=True)

party_vote_totals
```

```python
df_party = party_vote_totals

# คำนวณคะแนนเฉลี่ยต่อ ส.ส.บัญชีรายชื่อ 1 คน
average_vote_per_mp = df_party['total_party_list_vote'].sum() / 100

# คำนวณจำนวน ส.ส. แบบบัญชีรายชื่อที่แต่ละพรรคได้รับ
df_party['mp_count'] = df_party['total_party_list_vote'] // average_vote_per_mp

# กรณีส.ส.บัญชีรายชื่อยังได้ไม่ครบ 100 คน
# ให้พรรคการเมืองที่มีผลลัพธ์ที่เป็นเศษมากที่สุดได้รับส.ส.บัญชีรายชื่อเพิ่ม 1 คนเรียงตามลำดับ จนครบ 100 คน
remaining_seats = 100 - df_party['mp_count'].sum()
added_party_indices = []  # สร้างรายการเพื่อติดตามพรรคที่ได้รับการเพิ่มแล้ว
while remaining_seats > 0:
    # หาพรรคที่มีผลลัพธ์เศษมากที่สุดแต่ยังไม่ได้รับการเพิ่ม
    max_fractional_index = (df_party['total_party_list_vote'] % average_vote_per_mp)
    max_fractional_index = max_fractional_index.idxmax()
    while max_fractional_index in added_party_indices:
        # หาพรรคถัดไปที่มีผลลัพธ์เศษมากที่สุดแต่ยังไม่ได้รับการเพิ่ม
        df_party.at[max_fractional_index, 'total_party_list_vote'] = 0
        max_fractional_index = (df_party['total_party_list_vote'] % average_vote_per_mp)
        max_fractional_index = max_fractional_index.idxmax()
    df_party.at[max_fractional_index, 'mp_count'] += 1
    added_party_indices.append(max_fractional_index)
    remaining_seats -= 1


df_party[['name', 'mp_count']]
```

```python
papa = rank_counts.merge(df_party,how='left',on='name')
```

```python
papa2 =papa[['name','raw_score','mp_count']]
papa2['final_score']=papa2['raw_score']+papa2['mp_count']
papa2
```

# อำเภอที่หยุดนับ/นับ (2)

```python
import pandas as pd

# Merge result_constituencies_status with info_constituency on 'cons_id'
merged_df11 = result_constituencies_status[['cons_id', 'pause_report']].merge(
    info_constituency[['cons_id', 'zone']], on='cons_id', how='left'
)

# Merge merged_df1 with result_constituencies_Candidate on 'cons_id'
merged_df22 = merged_df11.merge(
    result_constituencies_Candidate[['cons_id', 'party_id']], on='cons_id', how='left'
)

# Merge merged_df2 with info_party_overview on 'party_id'
final_merged_df = merged_df22.merge(
    info_party_overview[['id']], left_on='party_id', right_on='id', how='left'
)

# Replace 'True' with 'หยุดนับคะแนน' and 'False' with 'ไม่หยุดนับคะแนน' in the 'pause_report' column
final_merged_df['pause_report'] = final_merged_df['pause_report'].replace(True, 'หยุดนับคะแนน')
final_merged_df['pause_report'] = final_merged_df['pause_report'].replace(False, 'ไม่หยุดนับคะแนน')

# Separate the DataFrame into two based on 'pause_report' values
pause_report_true_df = final_merged_df[final_merged_df['pause_report'] == 'หยุดนับคะแนน']
pause_report_false_df = final_merged_df[final_merged_df['pause_report'] == 'ไม่หยุดนับคะแนน']

# Remove duplicates from the 'zone' column
pause_report_true_df = pause_report_true_df.drop_duplicates(subset=['zone'])
pause_report_false_df = pause_report_false_df.drop_duplicates(subset=['zone'])

pause_report_true_df.to_csv('nostop.csv', index=True)
pause_report_false_df.to_csv('stop.csv', index=False)

```
